### Exercise #2 <br>
<p>Create a function that counts how many distinct words are in the string below, then outputs a dictionary with the words as the key and the value as the amount of times that word appears in the string.<br>Should output:<br>{'a': 5,<br>
 'abstract': 1,<br>
 'an': 3,<br>
 'array': 2, ... etc...</p>

```python
a_text = 'In computing, a hash table hash map is a data structure which implements an associative array abstract data type, a structure that can map keys to values. A hash table uses a hash function to compute an index into an array of buckets or slots from which the desired value can be found'

def organize(a_text):
    words = []
    words = a_text.split()
    repeat = [words.count(i) for i in words]
    myDict = dict(zip(words, repeat))
    print(myDict)
organize(a_text)
```

## Exercise #3

Write a program to implement a Linear Search Algorithm. Also in a comment, write the Time Complexity of the following algorithm.

#### Hint: Linear Searching will require searching a list for a given number. 

```python
# write a function
# input: a list, a thing to find in the list
# output: the index location of the thing in the list, or a message if the thing isn't in the list
# goal is to write the function such that it runs in linear O(n) time
# (don't sort)

b_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

def search(array, target): 
    left = 0
    right = len(array)-1
    target = target.lower()
    while left <= right:
        index = (left + right)//2
        potentialMatch = array[index]
        if target == potentialMatch:
            return f'The index of {target} is {index}.'
        elif target < potentialMatch:
            right = index-1
        else: # target > potentialMatch
            left = index+1
    return f'{target} is not in the list'

search(b_list, 'K')
```
